```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets

from IPython.display import display
```

```python
x = np.linspace(0, 10, 1000)

def plot_seno(freq=1.0):
    y = np.sin(2 * np.pi * freq * x)

    plt.figure(figsize=(10, 4))
    plt.plot(x, y)
    plt.grid(True)
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Amplitud")
    plt.title(f"Seno con frecuencia = {freq:.2f} Hz")
    plt.show()

widgets.interact(
    plot_seno,
    freq=widgets.FloatSlider(
        value=1.0,
        min=0.1,
        max=10.0,
        step=0.1,
        description="Frecuencia"
    )
)
```

```python
df = pd.DataFrame({
    "t": np.linspace(0, 20, 2000),
    "ax": np.random.randn(2000),
    "ay": np.random.randn(2000),
    "az": np.random.randn(2000),
})
```

```python
def plot_columna(columna="ax"):
    plt.figure(figsize=(12, 4))
    plt.plot(df["t"], df[columna])
    plt.grid(True)
    plt.xlabel("Tiempo [s]")
    plt.ylabel(columna)
    plt.title(f"Señal: {columna}")
    plt.show()

widgets.interact(
    plot_columna,
    columna=widgets.Dropdown(
        options=["ax", "ay", "az"],
        value="ax",
        description="Variable"
    )
)
```

```python
def plot_rango(t_ini=0.0, t_fin=10.0, columna="ax"):
    data = df[(df["t"] >= t_ini) & (df["t"] <= t_fin)]

    plt.figure(figsize=(12, 4))
    plt.plot(data["t"], data[columna])
    plt.grid(True)
    plt.xlabel("Tiempo [s]")
    plt.ylabel(columna)
    plt.title(f"{columna} entre {t_ini:.2f} s y {t_fin:.2f} s")
    plt.show()

widgets.interact(
    plot_rango,
    t_ini=widgets.FloatSlider(value=0, min=0, max=20, step=0.1, description="t_ini"),
    t_fin=widgets.FloatSlider(value=10, min=0, max=20, step=0.1, description="t_fin"),
    columna=widgets.Dropdown(options=["ax", "ay", "az"], value="ax", description="Variable")
)
```

```python
df_ins = pd.DataFrame({
    "t": np.linspace(0, 100, 1000),
    "px": np.cumsum(np.random.randn(1000) * 0.01),
    "py": np.cumsum(np.random.randn(1000) * 0.01),
})
```

```python
def plot_trayectoria(t_actual=10.0):
    data = df_ins[df_ins["t"] <= t_actual]

    plt.figure(figsize=(6, 6))
    plt.plot(data["px"], data["py"], label="Trayectoria")
    plt.scatter(data["px"].iloc[0], data["py"].iloc[0], label="Inicio")
    plt.scatter(data["px"].iloc[-1], data["py"].iloc[-1], label="Actual")

    plt.axis("equal")
    plt.grid(True)
    plt.xlabel("X [m]")
    plt.ylabel("Y [m]")
    plt.title(f"Trayectoria hasta t = {t_actual:.2f} s")
    plt.legend()
    plt.show()

widgets.interact(
    plot_trayectoria,
    t_actual=widgets.FloatSlider(
        value=10,
        min=df_ins["t"].min(),
        max=df_ins["t"].max(),
        step=0.5,
        description="Tiempo"
    )
)
```

```python
columna_widget = widgets.Dropdown(
    options=["ax", "ay", "az"],
    value="ax",
    description="Variable"
)

t_ini_widget = widgets.FloatSlider(
    value=0,
    min=0,
    max=20,
    step=0.1,
    description="t_ini"
)

t_fin_widget = widgets.FloatSlider(
    value=10,
    min=0,
    max=20,
    step=0.1,
    description="t_fin"
)

out = widgets.interactive_output(
    plot_rango,
    {
        "t_ini": t_ini_widget,
        "t_fin": t_fin_widget,
        "columna": columna_widget
    }
)

ui = widgets.VBox([
    columna_widget,
    t_ini_widget,
    t_fin_widget
])

display(ui, out)
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def plot_signal(df, time_col, signal_col, t_ini, t_fin):
    data = df[(df[time_col] >= t_ini) & (df[time_col] <= t_fin)]

    plt.figure(figsize=(12, 4))
    plt.plot(data[time_col], data[signal_col])
    plt.grid(True)
    plt.xlabel("Tiempo [s]")
    plt.ylabel(signal_col)
    plt.title(f"{signal_col} entre {t_ini:.2f}s y {t_fin:.2f}s")
    plt.show()


def create_signal_viewer(df, time_col="t"):
    signal_cols = [c for c in df.columns if c != time_col]

    signal_widget = widgets.Dropdown(
        options=signal_cols,
        value=signal_cols[0],
        description="Señal"
    )

    t_ini_widget = widgets.FloatSlider(
        value=float(df[time_col].min()),
        min=float(df[time_col].min()),
        max=float(df[time_col].max()),
        step=0.1,
        description="t_ini"
    )

    t_fin_widget = widgets.FloatSlider(
        value=min(float(df[time_col].min()) + 10, float(df[time_col].max())),
        min=float(df[time_col].min()),
        max=float(df[time_col].max()),
        step=0.1,
        description="t_fin"
    )

    out = widgets.interactive_output(
        lambda signal_col, t_ini, t_fin: plot_signal(
            df,
            time_col,
            signal_col,
            t_ini,
            t_fin
        ),
        {
            "signal_col": signal_widget,
            "t_ini": t_ini_widget,
            "t_fin": t_fin_widget
        }
    )

    ui = widgets.VBox([
        signal_widget,
        t_ini_widget,
        t_fin_widget
    ])

    display(ui, out)
```

```python
create_signal_viewer(df, time_col="t")
```
